```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\Lenovo\OneDrive\Desktop\deepLearning\DailyDelhiClimateTrain.csv')
```

```python
df.head()
```

```python
df.drop(['humidity','wind_speed','meanpressure'],axis=1,inplace=True)
```

```python
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],c='red')
```

```python
x1=df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewma']=x1
```

```python
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],c='red')
plt.plot(df['date'],df['ewma'],c='black',label='alpha=0.9')
plt.legend()
```
