```python
import tensorflow
from tensorflow import keras
from keras.layers import Conv2D,Flatten,Dense
from keras.models import Sequential
from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
model=Sequential()
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='valid',input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='valid'))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model1=Sequential()
model1.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='same',input_shape=(28,28,1)))
model1.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='same'))
model1.add(Flatten())
model1.add(Dense(128,activation='relu'))
model1.add(Dense(10,activation='softmax'))
```

```python
model1.summary()
```

```python
model1=Sequential()
model1.add(Conv2D(32,kernel_size=(3,3),activation='relu',strides=(2,2),padding='same',input_shape=(28,28,1)))
model1.add(Conv2D(32,kernel_size=(3,3),activation='relu',strides=(2,2),padding='same'))
model1.add(Conv2D(32,kernel_size=(3,3),activation='relu',strides=(2,2),padding='same'))
model1.add(Flatten())
model1.add(Dense(128,activation='relu'))
model1.add(Dense(10,activation='softmax'))
```
