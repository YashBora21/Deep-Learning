```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten

```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train[6]
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[4])
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train[0]
```

```python
model=Sequential()
#as our data have 28 row to make it in 1d we have to do flatten it
model.add(Flatten(input_shape=(28,28)))
#in the layer one we not have to give the inout as there we use flatten so it give it dirctly to dense
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))

#for multiclass we always use softmax
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=10,validation_split=0.2)
```

```python
history.history
```

```python
y_prob=model.predict(x_test)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
