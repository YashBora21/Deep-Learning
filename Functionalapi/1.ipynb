```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

x = Input(shape=(3,))
hidden1 = Dense(256, activation='relu')(x)
hidden2 = Dense(128,
 activation='relu')(hidden1)
hidden3 = Dense(64, activation='relu')(hidden2)

op1 = Dense(1, activation='linear')(hidden3)
op2 = Dense(1, activation='sigmoid')(hidden3)

model = Model(inputs=x, outputs=[op1, op2])
model.summary()
```
